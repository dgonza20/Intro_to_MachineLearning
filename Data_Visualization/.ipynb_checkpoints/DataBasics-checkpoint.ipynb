## Data Visualization

```python
# Splicing of total_UPDRS 
df1.loc[:, 'total_UPDRS']
```

```python
# Plotting total_UPDRS
plt.plot(df1.loc[:, 'total_UPDRS'])
```

```python
# Splicing of last column 
df1.loc[:, 'PPE']
```

```python
# Plotting the last feature/column of the data
plt.plot(df1.loc[:, 'PPE'])
```

```python
# Plotting a histogram of the last column
plt.hist(df1.loc[:, 'PPE'])
```

```python
# Splitting the data: Target
T1 = df1.loc[:, 'PPE'].copy()
T1
```

```python
# Putting the rest of the columns into a variable
X1 = df1.iloc[:, :-1].copy()
X1
```

```python
# Plotting with matlibplot: input vs target
fig = plt.figure(figsize=(16,12))
plt.clf() # Clear previous plt figure
for i in range(21):
    plt.subplot(5, 5, i+1) # Selects which subplot to plot to
    plt.plot(X1.iloc[:, i]) # Plots a given column
    plt.ylabel(X1.columns.values[i]) # Sets Y label
fig.tight_layout()
```

```python
# Plotting histogram of input vs output
fig = plt.figure(figsize=(16,12))
plt.clf() # Clear previous plt figure
for i in range(21):
    plt.subplot(5, 5, i+1) # Selects which subplot to plot to
    plt.hist(X1.iloc[:, i]) # Plots a given column
    plt.ylabel(X1.columns.values[i]) # Sets Y label
fig.tight_layout()
```

```python
# Concatinate the target and input
D1 = pd.concat([X1, T1], axis=1)
D1
```

```python
# Before applying log
T1.plot.hist()
```

```python
# Applying log
T1log = T1.copy()
T1log = np.log(T1log+1)
T1log
```

```python
# After applying log
T1log.plot.hist()
```
